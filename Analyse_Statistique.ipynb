**Exemple simple d'analyse statistique des données**

```python
import numpy as np
l=[3,2,10,18,1,4]
#On suit la boite à moustaches
print("Min:",np.min(l)) #Afficher le minimum des données
print("Q1",np.percentile(l,25)) #Afficher le quartile Q1=25% des données
print("Q2:",np.percentile(l,50))#Afficher le quartile Q2=50%=médiane des données
print("Q3:",np.percentile(l,75))#Afficher le quartile Q3=75% des données
print("Max:",np.max(l)) #Afficher le maximum des données 
print("Moyenne:",np.mean(l)) #Afficher la moyenne:somme/nbr termes
print("Variance:",np.var(l)) #Afficher la variance qui mesure à quel point les valeurs sont éloignées de la moyenne: var=(1/n)somme(xi-moy)^2 avec n=nbr de valeurs,xi:la valeur,moy:la moyenne 
print("Ecart type:",np.std(l)) #Afficher l'écart type qui mesure à quel point les valeurs sont éloignées ou rapprochées de la moyenne: std=racine(var)
print("Médiane:",np.median(l)) #Afficher la médiane
etendue=np.max(l)-np.min(l)
print("Etendue:",etendue)#Afficher l'etendue:Etendue=Max-Min
IQR=np.percentile(l,75)-np.percentile(l,25) #Mesurer la dispersion de 50% des données 
print("IQR:",IQR)
```
